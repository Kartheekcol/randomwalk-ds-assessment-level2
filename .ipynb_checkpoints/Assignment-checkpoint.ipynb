```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore, skew, kurtosis
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv('dataset.csv')
df
```

```python
#Missing or Incorrect Data
df.dropna(subset=['sex'],inplace=True)#Dropped the rows which has NaN in sex column as in other columns also it has NaN
missing_values = df.isnull().sum()
print(f"Missing Data before update: \n {missing_values}")
for col_name in df.columns:
    if df[col_name].dtype == "object":  
        df[col_name].fillna(df[col_name].mode()[0], inplace=True)#Object dtypes are replaced by mode
    else:  
        df[col_name].fillna(df[col_name].median(), inplace=True)#numerals are replaced by median
missing_values = df.isnull().sum()
print(f"Missing Data after update: \n{missing_values}")
```

```python
#Average body_mass_g for Gentoo penguins
gentoo = df[df['species'] == 'Gentoo']
print(f"Average body mass for Gentoo penguins: {gentoo['body_mass_g'].mean()}")
```

```python
# Check skewness and kurtosis for each species
for species in df['species'].unique():
    subset = df[df['species'] == species]
    for col in ['bill_length_mm', 'bill_depth_mm']:
        skewness = skew(subset[col])
        kurt = kurtosis(subset[col])
        print(f"Species: {species}, Feature: {col}, Skewness: {skewness:.2f}, Kurtosis: {kurt:.2f}")
```

```python
# Identify which features in the dataset have outliers
def detectoutliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return(df[(df[column] < lower_bound) | (df[column] > upper_bound)])
outliers = detectoutliers(df, 'body_mass_g')
sns.boxplot(x='species', y='body_mass_g', data=df)
plt.title("Outliers in Body Mass")
plt.show()
```

```python
df_pca = df.drop(columns=['species', 'island', 'sex'])
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_pca)
print("Explained variance ratio:", pca.explained_variance_ratio_)
```

```python
#bill_length_mm vs bill_depth_mm(7 different graphs)
fig, axs = plt.subplots(2, 4, figsize=(15, 10))
fig.suptitle('Bill Length vs Bill Depth Visualizations', fontsize=16)
# 1. Scatter Plot
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=df, ax=axs[0, 0])
axs[0, 0].set_title('Scatter Plot')
# 2. Joint Plot 
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=df, ax=axs[0, 1])
axs[0, 1].set_title('Scatter with Marginal KDE')
sns.kdeplot(data=df, x='bill_length_mm', ax=axs[1, 1], fill=True, color='blue', alpha=0.5)
sns.kdeplot(data=df, y='bill_depth_mm', ax=axs[0, 1], fill=True, color='blue', alpha=0.5)
# 3. Hexbin Plot
hb = axs[0, 2].hexbin(df['bill_length_mm'], df['bill_depth_mm'], gridsize=30, cmap='Blues', mincnt=1)
plt.colorbar(hb, ax=axs[0, 2], label='Counts')
axs[0, 2].set_xlabel('Bill Length (mm)')
axs[0, 2].set_ylabel('Bill Depth (mm)')
axs[0, 2].set_title('Hexbin Plot')
# 4. Box Plot
sns.boxplot(x='species', y='bill_length_mm', data=df, ax=axs[0, 3])
axs[0, 3].set_title('Box Plot of Bill Length by Species')
# 5. Violin Plot
sns.violinplot(x='species', y='bill_depth_mm', data=df, ax=axs[1, 0])
axs[1, 0].set_title('Violin Plot of Bill Depth by Species')
# 6. Pair Plot 
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=df, ax=axs[1, 1])
axs[1, 1].set_title('Pair Plot (Bill Length vs Depth)')
# 7. Histogram 
sns.histplot(df['bill_length_mm'], kde=True, bins=15, ax=axs[1, 2])
axs[1, 2].set_title('Histogram of Bill Length')
axs[1, 3].axis('off')
plt.tight_layout(rect=[0, 0, 1, 0.96]) 
plt.show()
```
